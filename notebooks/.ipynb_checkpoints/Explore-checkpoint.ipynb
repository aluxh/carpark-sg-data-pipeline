```python
import requests
import json
from datetime import datetime
import os
import logging
import psycopg2
import pandas as pd
import io
from pandas.io.json import json_normalize
```

```python
%load_ext sql
```

```python
parameters = { 'date' : '2019-08-09' }
results = requests.get("https://api.data.gov.sg/v1/environment/air-temperature?", parameters)
```

```python
if results.status_code == 200:
    json_data = results.json()
else:
    print("error in API call")
```

```python
weather_station = pd.DataFrame(json_data['metadata']['stations'])
weather_station_data = pd.concat([weather_station.id, weather_station.name, json_normalize(weather_station.location)], axis=1)
weather_station_data.rename(columns={'id':'station_id', 'name':'station_location', 'latitude':'station_latitude', 'longitude':'station_longitude'}, inplace=True)
weather_station_data
```

```python
x = json_normalize(json_data)
x = x.drop(columns=["metadata.reading_type", "metadata.reading_unit", "api_info.status"])
x.rename(columns={'items':'items', 'metadata.stations':'stations'}, inplace=True)
x.head()
```

```python
y = pd.DataFrame([(d, e) for tup in x.itertuples() for d in tup.items for e in tup.stations])
y.rename(columns={0:'timestamp', 1:'stations'}, inplace=True)
y.head()
```

```python
pd.DataFrame([(tup.timestamp, d) for tup in y.itertuples() for d in tup.timestamp])
```

```python
parameters = { 'date_time' : '2019-08-24T23:00:00' }
results = requests.get("https://api.data.gov.sg/v1/environment/rainfall?", parameters)
```

```python
if results.status_code == 200:
    json_data = results.json()
else:
    print("error in API call")
```

```python
json_data['items']
```

```python
rainfall = pd.DataFrame(json_data['items'])
```

```python
rainfall = pd.DataFrame([(tup.timestamp, d) for tup in rainfall.itertuples() for d in tup.readings])
```

```python
rainfall.rename(columns={0:'timestamp', 1:"readings"}, inplace=True)
```

```python
rainfall.head()
```

```python
rainfall = pd.concat([pd.to_datetime(rainfall['timestamp']), json_normalize(rainfall['readings'])], axis=1)
```

```python
rainfall['timestamp'] = [x.strftime("%Y-%m-%d %H:%M:%S") for x in rainfall['timestamp']]
```
